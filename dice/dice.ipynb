```python
from random import randint
import numpy as np
import matplotlib.pyplot as plt

tass = [0 , 0 , 0 , 0 , 0 , 0]
list1 =[0]
for i in range(1 , 10000):
    list1.append(0)
    
for i in range(10000):
    list1[i] = randint(1 , 6)
```

```python
tass[0] = list1.count(1)
tass[1] = list1.count(2)
tass[2] = list1.count(3)
tass[3] = list1.count(4)
tass[4] = list1.count(5)
tass[5] = list1.count(6)
print(tass)
```

```python
slices = tass
colors = ['red' , 'yellow' , 'green' , 'blue' , 'orange' , 'purple']
dice_numbers = ['1' , '2' , '3' , '4' , '5' , '6'] 
plt.pie(slices, labels = dice_numbers, colors = colors, startangle = 90, shadow = True, explode =(0.05, 0.05, 0.05, 0.05, 0.05, 0.05), radius = 1.2, autopct = '%1.1f%%')
plt.legend()
plt.show()
```
